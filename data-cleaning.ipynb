```python
#1. Load Data
df = pd.read_csv(r"C:\Users\burug\Downloads\marketing_campaign_dataset.csv")
```

```python
print("=" * 55)
print("STEP 1: RAW DATA OVERVIEW")
print("=" * 55)
print(f"Rows    : {df.shape[0]:,}")
print(f"Columns : {df.shape[1]}")
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 3 Rows:")
print(df.head(3))
```

```python
# 2.Understanding data types
print("\n" + "=" * 55)
print("STEP 2: DATA TYPES & NULLS")
print("=" * 55)
print(df.dtypes)
print(f"\nMissing values per column:")
print(df.isnull().sum())
```

```python
#3. Checking Duplicates
dupes = df.duplicated().sum()
print("\n" + "=" * 55)
print("STEP 3: DUPLICATES")
print("=" * 55)
print(f"Duplicate rows found: {dupes}")
if dupes > 0:
    df = df.drop_duplicates()
    print(f"Duplicates removed. New shape: {df.shape}")
else:
    print("No duplicates — data is clean.")
```

```python
#4.Standardize column name
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[^a-z0-9_]", "", regex=True)
)
print("\n" + "=" * 55)
print("STEP 4: CLEANED COLUMN NAMES")
print("=" * 55)
print(df.columns.tolist())
```

```python
#5. Standardize text values
text_cols = df.select_dtypes(include='object').columns.tolist()
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
 
print("\n" + "=" * 55)
print("STEP 5: UNIQUE VALUES IN KEY COLUMNS")
print("=" * 55)
for col in text_cols:
    print(f"  {col}: {df[col].unique()}")
```

```python
#6. convert numeric columns
print("\n" + "=" * 55)
print("STEP 6: NUMERIC CONVERSION")
print("=" * 55)
 
for col in df.columns:
    if df[col].dtype == object:
        try:
            cleaned = df[col].str.replace('%', '', regex=False).astype(float)
            df[col] = cleaned
            print(f"  Converted '{col}' from string/% to float")
        except (ValueError, AttributeError):
            pass
 
print("Numeric columns after conversion:")
print(df.select_dtypes(include='number').columns.tolist())
```

```python
print("\n" + "=" * 55)
print("STEP 7: DERIVED METRICS")
print("=" * 55)
 
# ROI
if 'roi' in df.columns:
    print("  ROI column exists — using as-is")
else:
    if 'revenue' in df.columns and 'cost' in df.columns:
        df['roi'] = ((df['revenue'] - df['cost']) / df['cost'] * 100).round(2)
        print("  Calculated ROI from revenue and cost")
 
# CAC
if 'cac' in df.columns:
    print("  CAC column exists — using as-is")
else:
    if 'cost' in df.columns and 'conversions' in df.columns:
        df['cac'] = (df['cost'] / df['conversions'].replace(0, np.nan)).round(2)
        print("  Calculated CAC from cost / conversions")
 
# Engagement Rate
if 'engagement_rate' in df.columns:
    print("  Engagement Rate column exists — using as-is")
else:
    if 'clicks' in df.columns and 'impressions' in df.columns:
        df['engagement_rate'] = (df['clicks'] / df['impressions'].replace(0, np.nan) * 100).round(4)
        print("  Calculated Engagement Rate from clicks / impressions")
 
# Conversion Rate
if 'conversion_rate' in df.columns:
    print("  Conversion Rate column exists — using as-is")
else:
    if 'conversions' in df.columns and 'clicks' in df.columns:
        df['conversion_rate'] = (df['conversions'] / df['clicks'].replace(0, np.nan) * 100).round(4)
        print("  Calculated Conversion Rate from conversions / clicks")
 
print("\nAll columns after deriving metrics:")
print(df.columns.tolist())
```

```python
print("\n" + "=" * 55)
print("STEP 8: SUMMARY STATISTICS")
print("=" * 55)
print(df.describe().round(2))
```

```python
print("\n" + "=" * 55)
print("STEP 9: RECORDS PER CHANNEL")
print("=" * 55)
channel_col = next((c for c in df.columns if 'channel' in c), None)
if channel_col:
    print(df[channel_col].value_counts())
else:
    print("No 'channel' column found — check Step 4 column names above.")
```

```python
output_path = r"C:\Users\burug\Downloads\marketing_campaign_cleaned.csv"
df.to_csv(output_path, index=False)
 
print("\n" + "=" * 55)
print("STEP 10: CLEANED FILE SAVED")
print("=" * 55)
print(f"Saved to: {output_path}")
print(f"Final shape: {df.shape[0]:,} rows x {df.shape[1]} columns")
print("\nData cleaning complete. Ready for SQL analysis.")
```
